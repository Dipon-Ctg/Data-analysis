## Q1: How many people have heart disease and how many people doesn't have heart disease?

```python
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind= 'bar',color=['orange', 'red'], figsize=(6,3));
```

```python
df.target.value_counts().plot(kind= 'bar',color=['orange', 'red'], figsize=(6,3));
plt.title('Heart Disease Value')
plt.xlabel('1= Disease,0=No Disease')
plt.ylabel('Amounts');
```

```python
df.target.value_counts().plot(kind='pie')
```

```python
df.target.value_counts().plot(kind='pie');                    #Using ';' for remove the lines
```

```python
df.target.value_counts().plot(kind='pie')
plt.legend(['Disease','No Disease']);
```

```python
df.target.value_counts().plot(kind='pie', figsize=(6,7))      #Increasing size so that there is no overlap
plt.legend(['Disease','No Disease']);
```

```python
df.target.value_counts().plot(kind='pie', figsize=(6,7))      #Labeling
plt.legend(['Disease','No Disease'])
plt.xlabel('1= Disease,0= No Disease');
```

## Q2: People of which sex has most heart disease?

```python
df.sex.value_counts().plot(kind='bar',color=['DarkGreen', 'Red']);
plt.title('Sex')
plt.xlabel('1= Male,0=Female')
plt.ylabel('Amounts');
```

```python
pd.crosstab(df.sex,df.target)
```
